# Demo of Engineering Calculations Using ipython Notebook

```python
from IPython.display import Image
Image(filename='Summary.png')
```

```python
x = []
y = []
z = []
import math
```

```python
T = 5 # N*m   4 ft*lbs = 
l = 22 # mm
r = 7.17/1000 # m
Fs = T/r ; print('tangential force, Fs=',Fs, 'N')
phi = math.radians(20) # pressure angle
Fr = Fs*math.tan(phi)
print('radial load Fr = ',Fr , ' and tangential load Fs = ' ,Fs)
# 1 lb = 4.4 N
```

- - - -

## Beam bending due to Normal Load P_1

```python
from IPython.display import Image
Image(filename='1.png')
```

```python
P1 = Fr*math.cos(math.pi/4) ; print(P1)
L = 90 #mm
E = 200.0e3 # MPa
b = 3.5 #mm
d = 6 #mm
I = (b*d**3) / 12.0
I = I*2  # two cantilevers
z.append( (P1*L**3) / (3*E*I) )
print('total vertical deflection due to normal load, z = ',z[0],'mm')
```

- - - -

## Beam extension due to axial Load P_2

```python
from IPython.display import Image
Image(filename='2.png')
```

```python
P2 = Fr*math.sin(math.pi/4) ; print(P2)
A = 2*b*d
y.append( P2*L / (A*E) )
print('total horizontal extension due to axial load, y = ',y[0],'mm')
```

- - - -

## Beam Rotation about axis due to Moment, M

```python
from IPython.display import Image
Image(filename='3.png')
```

```python
M = Fr*math.sin(math.pi/4)*l
z.append( (M*L**2) / (2*E*I) )
print('vertical deflection = ',z[1],'mm')
```

- - - -

## Beam Bending due to tangential force Fs

```python
from IPython.display import Image
Image(filename='4.png')
```

```python
L = 90 #mm
E = 200.0e3 # MPa
b = 6 ; d1 = 13 ; d2 = 5
I = (b / 12.0) * (d1**3.0 - d2**2)
z.append( (P1*L**3) / (3*E*I) )
print('total vertical deflection due to normal load, z = ',z[2],'mm')
```

- - - -

## Beam angular deflection due to torsion from tangential force 

```python
from IPython.display import Image
Image(filename='5.png')
```

```python
G = E / (2*(1+0.29))
T = l*Fs
a = 13/2.0 ; b = 6/2.0
K1 = a*b**3*(16/3-3.36*b/a*(1-b**4/(12*a**4)))
a = 5.0/2 ; b = 6/3.0
K2 = a*b**3*(16/3-3.36*b/a*(1-b**4/(12*a**4)))
K = K1-K2
Theta = (T*L)/(K*G)
print('rotation, theta = ',math.degrees(Theta),'degrees')
```

```python
delta = math.sqrt(sum(z)**2 + sum(y)**2)
print('total deflection in z = ',sum(z))
print('total deflection in y = ',sum(y))
print('total deflection = ',delta,'mm')
```
